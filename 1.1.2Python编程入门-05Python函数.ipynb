# 函数定义


```python
# 三角形面积公式
# 面积 = 底长 * 高长 / 2

def cal_triangle_area(bottom, height):
    return bottom * height / 2

print('底长为3，高长为4的三角形面积:', end='') #end表示不换行
print(cal_triangle_area(3, 4))


# 摄氏度到华氏度转换函数
# 公式：华氏度 = 摄氏度*9/5+32
def fahrenheit_converter(C):
     fahrenheit = C * 9/5 + 32
     return str(fahrenheit) + '˚F'

print('摄氏度是37，转换为华氏度是:', end='')
print(fahrenheit_converter(37))
```

```python
# 下列代码演示了函数如何返回多个值

# 返回用户信息，包括姓名，年龄，性别，婚姻状态
def get_user_info():
    return '张三', 18, '男', '未婚'

name, age, gender, marital_status = get_user_info()
print ('姓名:' + name)
print ('年龄:' + str(age))
print ('性别:' + gender)
print ('婚姻状态:' + marital_status)
```

# 函数参数


## 必选参数

```python
def cal_triangle_area(bottom, height):
    return bottom * height / 2

print(cal_triangle_area(10)) # 应该传递两个必选参数，实际中只传递了一个，报错 
```

## 默认参数

```python
def cal_triangle_area(bottom, height=3):
    return bottom * height / 2

print(cal_triangle_area(10)) # 使用默认参数3

print(cal_triangle_area(10, 4)) # 使用传递参数4
```

## 可变参数

```python
def printinfo( arg1, *vartuple ):
    print ("输出: ")
    print (arg1)
    print (vartuple)
 
# 调用printinfo函数
printinfo( 70, 60, 50,40) # 参数60和50都是可变参数
```

## 关键字参数

```python
# 关键字参数演示
def person(name, age, **kw):
    print('name:', name, 'age:', age, 'other:', kw)
    
# 函数person除了必选参数name和age外，还接受关键字参数kw。在调用该函数时，可以只传入必选参数

person('ZhiWen', 2)
person('ZhiWen', 2, city='Beijing')
person('David', 30, gender='F', job='Engineer')
```

# return语句

```python
def judge(annual_income):
    if annual_income > 0 and annual_income<= 100000:
        print('特困家庭')
    elif annual_income > 100000 and annual_income <= 200000:
        print('困难家庭')
    elif annual_income > 200000 and annual_income<= 350000:
         print('贫困家庭')
    elif annual_income > 200000 and annual_income<= 350000:
         print('困难家庭')
    elif annual_income > 350000 and annual_income<= 600000:
         print('贫穷家庭')
    elif annual_income > 600000 and annual_income<= 800000:
         print('小康家庭')
    elif annual_income > 800000 and annual_income<= 2000000:
         print('中产家庭')
    elif annual_income > 2000000 and annual_income<= 5000000:
         print('富裕家庭')
    else:
         print('富人家庭')

annual_income = int(input("请输你家的年收入总额: "))
judge(annual_income)

if None == judge(annual_income):
    print('函数返回None')
```
